```python
# ============================================
# SWAT FLOOD SEGMENTATION - U-NET FIXED 2025
# ============================================

# STEP 1 - INSTALL AND IMPORT
!pip install rasterio --quiet
!pip install imbalanced-learn --quiet

import numpy as np
import matplotlib.pyplot as plt
import rasterio
from skimage.transform import resize
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
print("All libraries imported!")

# STEP 2 - MOUNT DRIVE
from google.colab import drive
drive.mount('/content/drive')

# STEP 3 - SET PATH
path = '/content/drive/MyDrive/swat_flood_2025/'

# STEP 4 - READ ALL FILES
with rasterio.open(path + 'Swat_SAR_Before_2025.tif') as src:
    sar_before = src.read(1).astype(float)
with rasterio.open(path + 'Swat_SAR_During_2025.tif') as src:
    sar_during = src.read(1).astype(float)
with rasterio.open(path + 'Swat_Flood_Extent_Fixed.tif') as src:
    flood_label = src.read(1).astype(float)
with rasterio.open(path + 'Swat_DEM_2025.tif') as src:
    dem = src.read(1).astype(float)
with rasterio.open(path + 'Swat_Slope_2025.tif') as src:
    slope = src.read(1).astype(float)
with rasterio.open(path + 'Swat_Rainfall_CHIRPS_2025.tif') as src:
    rainfall = src.read(1).astype(float)
print("All files loaded!")
print("Original shape:", sar_before.shape)

# STEP 5 - RESIZE TO LARGER SIZE
# Use 512x512 for more patches
target = (512, 512)
sar_before  = resize(sar_before,  target)
sar_during  = resize(sar_during,  target)
flood_label = resize(flood_label, target)
dem         = resize(dem,         target)
slope       = resize(slope,       target)
rainfall    = resize(rainfall,    target)
print("Resized to:", target)

# STEP 6 - NORMALIZE
def normalize(data):
    data = np.nan_to_num(data, nan=0.0)
    if data.max() != data.min():
        return (data - data.min()) / (data.max() - data.min())
    return data

sar_before  = normalize(sar_before)
sar_during  = normalize(sar_during)
dem         = normalize(dem)
slope       = normalize(slope)
rainfall    = normalize(rainfall)
flood_label = (flood_label > 0.5).astype(float)
flood_label = np.nan_to_num(flood_label, nan=0.0)

print("Flood pixels in label:", flood_label.sum())
print("No flood pixels      :", (flood_label==0).sum())
print("Normalization complete!")

# STEP 7 - VISUALIZE DATA
fig, axes = plt.subplots(2, 3, figsize=(16,10))
axes[0,0].imshow(sar_before,  cmap='gray')
axes[0,0].set_title('SAR Before Flood')
axes[0,0].axis('off')
axes[0,1].imshow(sar_during,  cmap='gray')
axes[0,1].set_title('SAR During Flood')
axes[0,1].axis('off')
axes[0,2].imshow(flood_label, cmap='Blues')
axes[0,2].set_title('Flood Label')
axes[0,2].axis('off')
axes[1,0].imshow(dem,         cmap='terrain')
axes[1,0].set_title('DEM')
axes[1,0].axis('off')
axes[1,1].imshow(slope,       cmap='YlOrRd')
axes[1,1].set_title('Slope')
axes[1,1].axis('off')
axes[1,2].imshow(rainfall,    cmap='Blues')
axes[1,2].set_title('Rainfall')
axes[1,2].axis('off')
plt.suptitle('Swat Flood Data 2025', fontsize=16)
plt.tight_layout()
plt.savefig('swat_data_unet.png', dpi=150)
plt.show()

# STEP 8 - CREATE PATCHES WITH SMALLER SIZE
def create_patches(images, label, patch_size=64, stride=16):
    patches_x = []
    patches_y = []
    flood_patches_x = []
    flood_patches_y = []
    h, w = label.shape

    for i in range(0, h - patch_size + 1, stride):
        for j in range(0, w - patch_size + 1, stride):
            patch = np.stack([
                img[i:i+patch_size, j:j+patch_size]
                for img in images
            ], axis=-1)
            label_patch = label[
                i:i+patch_size,
                j:j+patch_size
            ]

            # Separate flood and no flood patches
            if label_patch.sum() > 10:
                flood_patches_x.append(patch)
                flood_patches_y.append(label_patch)
            else:
                patches_x.append(patch)
                patches_y.append(label_patch)

    print(f"Flood patches    : {len(flood_patches_x)}")
    print(f"No flood patches : {len(patches_x)}")

    # Balance patches
    # Use all flood patches
    # Sample equal no flood patches
    n_flood = len(flood_patches_x)
    if n_flood == 0:
        print("WARNING: No flood patches found!")
        print("Using all patches...")
        all_x = patches_x
        all_y = patches_y
    else:
        # Take 3x no flood patches
        import random
        random.seed(42)
        n_noflood = min(n_flood * 3, len(patches_x))
        selected_noflood = random.sample(
            list(zip(patches_x, patches_y)),
            n_noflood
        )
        noflood_x = [p[0] for p in selected_noflood]
        noflood_y = [p[1] for p in selected_noflood]

        all_x = flood_patches_x + noflood_x
        all_y = flood_patches_y + noflood_y

    return np.array(all_x), np.array(all_y)

images = [
    sar_before,
    sar_during,
    dem,
    slope,
    rainfall
]

X_patches, y_patches = create_patches(
    images,
    flood_label,
    patch_size=64,
    stride=16
)

y_patches = np.expand_dims(y_patches, axis=-1)

print("\nTotal patches    :", X_patches.shape[0])
print("Patch shape      :", X_patches.shape[1:])
print("Label shape      :", y_patches.shape[1:])
print("Flood pixels     :", y_patches.sum())

# STEP 9 - SPLIT DATA
X_train, X_test, y_train, y_test = train_test_split(
    X_patches, y_patches,
    test_size=0.2,
    random_state=42
)
print("Train patches:", X_train.shape[0])
print("Test patches :", X_test.shape[0])

# STEP 10 - BUILD U-NET
def unet_model(input_shape=(64, 64, 5)):
    inputs = keras.Input(shape=input_shape)

    # ENCODER
    c1 = layers.Conv2D(16, 3, activation='relu', padding='same')(inputs)
    c1 = layers.Conv2D(16, 3, activation='relu', padding='same')(c1)
    p1 = layers.MaxPooling2D(2)(c1)
    p1 = layers.Dropout(0.1)(p1)

    c2 = layers.Conv2D(32, 3, activation='relu', padding='same')(p1)
    c2 = layers.Conv2D(32, 3, activation='relu', padding='same')(c2)
    p2 = layers.MaxPooling2D(2)(c2)
    p2 = layers.Dropout(0.1)(p2)

    c3 = layers.Conv2D(64, 3, activation='relu', padding='same')(p2)
    c3 = layers.Conv2D(64, 3, activation='relu', padding='same')(c3)
    p3 = layers.MaxPooling2D(2)(c3)
    p3 = layers.Dropout(0.2)(p3)

    # BOTTLENECK
    c4 = layers.Conv2D(128, 3, activation='relu', padding='same')(p3)
    c4 = layers.Conv2D(128, 3, activation='relu', padding='same')(c4)

    # DECODER
    u5 = layers.Conv2DTranspose(64, 2, strides=2, padding='same')(c4)
    u5 = layers.concatenate([u5, c3])
    c5 = layers.Conv2D(64, 3, activation='relu', padding='same')(u5)
    c5 = layers.Conv2D(64, 3, activation='relu', padding='same')(c5)
    c5 = layers.Dropout(0.2)(c5)

    u6 = layers.Conv2DTranspose(32, 2, strides=2, padding='same')(c5)
    u6 = layers.concatenate([u6, c2])
    c6 = layers.Conv2D(32, 3, activation='relu', padding='same')(u6)
    c6 = layers.Conv2D(32, 3, activation='relu', padding='same')(c6)
    c6 = layers.Dropout(0.1)(c6)

    u7 = layers.Conv2DTranspose(16, 2, strides=2, padding='same')(c6)
    u7 = layers.concatenate([u7, c1])
    c7 = layers.Conv2D(16, 3, activation='relu', padding='same')(u7)
    c7 = layers.Conv2D(16, 3, activation='relu', padding='same')(c7)
    c7 = layers.Dropout(0.1)(c7)

    outputs = layers.Conv2D(1, 1, activation='sigmoid')(c7)
    model = keras.Model(inputs, outputs)
    return model

model = unet_model(input_shape=(64, 64, 5))
model.summary()

# STEP 11 - COMPILE WITH WEIGHTED LOSS
def weighted_bce(y_true, y_pred):
    weight = tf.where(
        tf.equal(y_true, 1),
        10.0,
        1.0
    )
    bce = tf.keras.losses.binary_crossentropy(
        y_true, y_pred
    )
    return tf.reduce_mean(bce * tf.squeeze(weight, -1))

model.compile(
    optimizer='adam',
    loss=weighted_bce,
    metrics=[
        'accuracy',
        tf.keras.metrics.Recall(),
        tf.keras.metrics.Precision()
    ]
)
print("Model compiled with weighted loss!")

# STEP 12 - TRAIN MODEL
history = model.fit(
    X_train, y_train,
    epochs=20,
    batch_size=8,
    validation_split=0.2,
    verbose=1,
    callbacks=[
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=5,
            restore_best_weights=True
        )
    ]
)
print("Training complete!")

# STEP 13 - PRINT EPOCH RESULTS
print("\nEpoch Results:")
print("-" * 70)
for i in range(len(history.history['accuracy'])):
    print(
        f"Epoch {i+1:2d} | "
        f"Acc: {history.history['accuracy'][i]:.4f} | "
        f"Val Acc: {history.history['val_accuracy'][i]:.4f} | "
        f"Loss: {history.history['loss'][i]:.4f} | "
        f"Val Loss: {history.history['val_loss'][i]:.4f}"
    )

# STEP 14 - PLOT RESULTS
fig, axes = plt.subplots(1, 2, figsize=(14,5))
axes[0].plot(history.history['accuracy'],
             label='Train', color='blue', marker='o')
axes[0].plot(history.history['val_accuracy'],
             label='Validation', color='orange', marker='o')
axes[0].set_title('U-Net Accuracy')
axes[0].set_xlabel('Epoch')
axes[0].legend()
axes[0].grid(True)
axes[1].plot(history.history['loss'],
             label='Train', color='blue', marker='o')
axes[1].plot(history.history['val_loss'],
             label='Validation', color='orange', marker='o')
axes[1].set_title('U-Net Loss')
axes[1].set_xlabel('Epoch')
axes[1].legend()
axes[1].grid(True)
plt.suptitle('U-Net Training — Swat Flood 2025',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('unet_training_results.png', dpi=150)
plt.show()

# STEP 15 - EVALUATE
results = model.evaluate(X_test, y_test, verbose=0)
print(f"\nTest Accuracy  : {results[1]*100:.2f}%")
print(f"Test Loss      : {results[0]:.4f}")
print(f"Test Recall    : {results[2]:.4f}")
print(f"Test Precision : {results[3]:.4f}")

# STEP 16 - PREDICTIONS
predictions = model.predict(X_test, verbose=0)
pred_binary = (predictions > 0.5).astype(int)

# Show sample predictions
fig, axes = plt.subplots(3, 4, figsize=(16,12))
for i in range(4):
    axes[0,i].imshow(X_test[i,:,:,0], cmap='gray')
    axes[0,i].set_title(f'SAR {i+1}')
    axes[0,i].axis('off')
    axes[1,i].imshow(y_test[i,:,:,0], cmap='Blues')
    axes[1,i].set_title(f'Actual {i+1}')
    axes[1,i].axis('off')
    axes[2,i].imshow(pred_binary[i,:,:,0], cmap='Blues')
    axes[2,i].set_title(f'Predicted {i+1}')
    axes[2,i].axis('off')
plt.suptitle('U-Net Predictions — Swat 2025',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('unet_predictions.png', dpi=150)
plt.show()

# STEP 17 - CONFUSION MATRIX
y_true = y_test.flatten().astype(int)
y_pred = pred_binary.flatten().astype(int)

print("\nUnique values in y_true:", np.unique(y_true))
print("Unique values in y_pred:", np.unique(y_pred))

cm = confusion_matrix(y_true, y_pred, labels=[0,1])
plt.figure(figsize=(8,6))
sns.heatmap(cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['No Flood','Flood'],
    yticklabels=['No Flood','Flood']
)
plt.title('Confusion Matrix — U-Net Swat 2025')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.savefig('unet_confusion_matrix.png', dpi=150)
plt.show()

print("\nClassification Report:")
print(classification_report(
    y_true, y_pred,
    labels=[0,1],
    target_names=['No Flood','Flood']
))

# STEP 18 - FLOOD MAP
print("Creating flood map...")
full_image = np.stack([
    sar_before,
    sar_during,
    dem,
    slope,
    rainfall
], axis=-1)
full_image = np.expand_dims(full_image, axis=0)
flood_map = model.predict(full_image, verbose=0)
flood_map = flood_map[0,:,:,0]

fig, axes = plt.subplots(1, 3, figsize=(18,6))
axes[0].imshow(sar_during, cmap='gray')
axes[0].set_title('SAR During Flood')
axes[0].axis('off')
axes[1].imshow(flood_label, cmap='Blues')
axes[1].set_title('Actual Flood')
axes[1].axis('off')
axes[2].imshow(flood_map, cmap='Blues')
axes[2].set_title('U-Net Predicted')
axes[2].axis('off')
plt.suptitle('Swat Flood Map 2025 — U-Net',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('unet_flood_map.png', dpi=150)
plt.show()

# STEP 19 - SAVE
model.save('swat_unet_model.keras')
print("Model saved!")
print("=" * 40)
print("U-NET COMPLETE!")
print("=" * 40)
```
